```python
# 数据集中出现了部分序号重复的记录，考虑到封面图片的问题，因此将全部的重复记录删除
df.drop_duplicates(subset=['序号'],keep=False,inplace=True)

## 1.书籍类型
book_types=['历史知识', '科学知识', '节日民俗', '自然生物', '数学知识', '线条色彩','安全教育', '传统文化', '成长道理', '健康问题', '体育运动', '习惯养成', '艺术修养', '趣味解密', '哲理思考','情感体验']
df[book_types]=df[book_types].applymap(lambda x: 1 if x==1 else 0)
for i in df.index:
    for type in book_types:
        if df.loc[i,type]==1:
            df.loc[i,'book_type']=type
for i in df.index:
    if df.loc[i,'book_type'] not in book_types:
        df.loc[i,'book_type']='情感体验'


## 2.查找是否有相应的封面图片
import os
#获取目录下文件名清单
files=os.listdir("./cover_pic")
pic_index=[] 
for filename in files:
    portion=os.path.splitext(filename)
    pic_index.append(int(portion[0]))
for i in df.index:
    if df.loc[i,'序号'] not in pic_index:
        df.loc[i,'have_cover_pic']=0
    else:
        df.loc[i,'have_cover_pic']=1

## 3.主角归类，全部有主角，分类为动物、植物、人物、食物、交通工具
animals_list=['猫', '狗', '鼠', '兔子', '龙猫', '刺猬', '狐狸', '松鼠', '猴子', '猪','熊', '牛', '马', '鹿', '狼', '羊', '恐龙', '大象', '骆驼', '狮子', '老虎', '豹子', '长颈鹿','蛇', '乌龟', '壁虎', '青蛙', '鳄鱼', '变色龙', '蝉', '蜗牛', '瓢虫', '蚂蚁', '蜻蜓', '蝴蝶','蜘蛛', '蚯蚓', '蜜蜂', '独角仙', '萤火虫', '毛毛虫', '鸭', '鸡', '鸟', '孔雀', '虾', '蟹','鱼', '海豚', '鲸鱼', '鲨鱼']
plants_list=['向日葵', '蒲公英', '豆子', '香蕉', '花朵', '仙人掌', '葡萄','萝卜', '树', '草', '橘子', '苹果', '草莓', '土豆', '番茄', '豆荚', '茄子']
figures_list=['公主', '王子','国王', '王后', '和尚', '道士', '医生', '驾驶员', '老人', '警察', '售货员', '大将军', '老师','猎人', '探险家', '宇航员', '盗贼', '海盗', '管理员', '科学家', '文学家']
food_list=['披萨', '汉堡', '奶酪', '冰激凌', '包子', '饺子','火锅', '月饼', '粽子', '豆腐', '汤圆', '甜品', '巧克力', '糖果', '肉丸','甜甜圈']
vehicles_list=['校车', '公交巴士','挖掘机', '小汽车', '卡车', '火车', '船', '消防车', '马车', '自行车', '飞机', '压路车', '缆车','有轨交通', '潜艇']
others_list=['少数民族', '家庭成员', '儿童', '历史人物', '太阳', '地球', '月亮', '星星', '火箭', '卫星','彗星', '宇宙飞船', '佛', '神仙', '天使', '上帝', '鬼魂', '精灵', '超人', '小矮人', '巨人', '龙', '凤', '魔法师', '怪兽', '独角兽', '外星人', '奥特曼', '圣诞老人', '僵尸', '鞋靴', '雪人', '机器人','牙齿', '蜡笔', '雨伞', '房子', '背包', '水', '风', '土地', '家用电器', '衣物', '云朵','厨具', '图形', '气球', '玩具', '音乐', '工具', '字母汉字']

df[animals_list]=df[animals_list].applymap(lambda x: 1 if x==1 else 0)
df[plants_list]=df[plants_list].applymap(lambda x: 1 if x==1 else 0)
df[figures_list]=df[figures_list].applymap(lambda x: 1 if x==1 else 0)
df[food_list]=df[food_list].applymap(lambda x: 1 if x==1 else 0)
df[vehicles_list]=df[vehicles_list].applymap(lambda x: 1 if x==1 else 0)
df[others_list]=df[others_list].applymap(lambda x: 1 if x==1 else 0)
for i in df.index:
    # 动物主角
    animals=""
    for animal in animals_list:
        if df.loc[i,animal]==1:
            if len(animals)==0:
                animals=animal
            else:
                animals=animals+","+animal
    df.loc[i,'main_character_animals']=animals
    # 植物主角
    plants=""
    for plant in plants_list:
        if df.loc[i,plant]==1:
            if len(plants)==0:
                plants=plant
            else:
                plants=plants+","+plant
    df.loc[i,'main_character_plants']=plants
    # 人物主角
    figures=""
    for figure in figures_list:
        if df.loc[i,figure]==1:
            if len(figures)==0:
                figures=figure
            else:
                figures=figures+","+figure
    df.loc[i,'main_character_figures']=figures
    # 食物主角
    foods=""
    for food in food_list:
        if df.loc[i,food]==1:
            if len(foods)==0:
                foods=food
            else:
                foods=foods+","+food
    df.loc[i,'main_character_foods']=foods
    # 交通工具主角
    vehicles=""
    for vehicle in vehicles_list:
        if df.loc[i,vehicle]==1:
            if len(vehicles)==0:
                vehicles=vehicle
            else:
                vehicles=vehicles+","+vehicle
    df.loc[i,'main_character_vehicles']=vehicles
    # 其他主角
    others=""
    for other in others_list:
        if df.loc[i,other]==1:
            if len(others)==0:
                others=other
            else:
                others=others+","+other
    df.loc[i,'main_character_others']=others


# 4.主角特点
protagonist_features=['外貌', '性格', '情绪', '职业', '身份', '能力']
df[protagonist_features]=df[protagonist_features].applymap(lambda x: 1 if x==1 else 0)
for i in df.index:
    features=""
    for feature in protagonist_features:
        if df.loc[i,feature]==1:
            if len(features)==0:
                features=feature
            else:
                features=features+","+feature
    df.loc[i,'protagonist_features']=features

# 5.主角之间关系
protagonist_relations=['家庭关系', '朋友关系', '宠物关系', '敌人关系', '师生关系', '同学关系', '邻居关系','群体关系']
df[protagonist_relations]=df[protagonist_relations].applymap(lambda x: 1 if x==1 else 0)
for i in df.index:
    relations=""
    for relation in protagonist_relations:
        if df.loc[i,relation]==1:
            if len(relations)==0:
                relations=relation
            else:
                relations=relations+","+relation
    df.loc[i,'protagonist_relations']=relations

# 6.时间因素
time_factors=['春', '夏', '秋', '冬', '四季轮转', '早晨', '中午', '傍晚', '夜晚','朝夕轮转', '雨', '雪', '打雷闪电', '大风', '彩虹', '节日', '生日']
df[time_factors]=df[time_factors].applymap(lambda x: 1 if x==1 else 0)
for i in df.index:
    time=""
    for factor in time_factors:
        if df.loc[i,factor]==1:
            if len(time)==0:
                time=factor
            else:
                time=time+","+factor
    df.loc[i,'time_factors']=time

# 7.地点因素
location_factors=['家', '学校', '城镇','饭店', '商店', '乡村', '自然', '农牧场', '宇宙', '水域', '医院', '公园', '动物园', '图书馆','博物馆', '海洋馆', '异世界', '天空', '沙滩', '地下', '城堡', '花园', '梦境', '实验室', '沙漠','交通工具']
df[location_factors]=df[location_factors].applymap(lambda x: 1 if x==1 else 0)
for i in df.index:
    location=""
    for factor in location_factors:
        if df.loc[i,factor]==1:
            if len(location)==0:
                location=factor
            else:
                location=location+","+factor
    df.loc[i,'location_factors']=location

# 8.标志性物品
df['have_iconic_items']=df['有无标志性物品'].apply(lambda x: 1 if x==1 else 0)
```

```python
# 1.对于一本给定的图画书绘本，根据其作者查找同系列书籍，使用给定图书的序号

# 查看书籍是否有相同的前缀，若有则说明是同一系列的书籍
def have_same_prefix(str1,str2):
    if str2.startswith(str1[:2]):
        return True
    else:
        return False
    
def find_series_books(serial_number):
    series_list=[]
    same_author_list=[]
    serial_number=int(serial_number)
    author=df.loc[df['serial_number']==serial_number,'author'].values[0]
    if len(df[df['author']==author].index)==1:
        return [-1],[-1]
    book_name=df.loc[df['serial_number']==serial_number,'book_name'].values[0]

    for i in df[df['author']==author].index:
        if int(df.loc[i,'serial_number'])!=serial_number:
            if df.loc[i,'is_series']==1 and have_same_prefix(book_name,str(df.loc[i,'book_name']))==True:
                series_list.append(df.loc[i,'serial_number'])
            else:
                same_author_list.append(df.loc[i,'serial_number'])
        else:
            continue
    return series_list,same_author_list

```
